```
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img = cv.imread('/content/drive/MyDrive/kuliyah/PCVK/kitten01.jpg')
plt.imshow(img)
```

```
# Red green
imgt = img
imgt[:,:,0] = np.zeros([imgt.shape[0], imgt.shape[1]])
plt.imshow(imgt)
```

```
# green blue
imgt1 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
imgt1[:,:,2] = np.zeros([imgt1.shape[0], imgt1.shape[1]])
plt.imshow(imgt1)
```

```
# 2. Tampilkan image baris ke 10-100, kolom 10-100!
imgt = img[10:100,10:100]
plt.imshow(imgt)
```

```
# 3. Tampilkan image baris ke 5-30, semua kolom, channel Green saja
plt.imshow(img[5:30])
```

```
# 4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
img3 = cv.resize(cv.cvtColor(img,cv.COLOR_BGR2RGB),(850,850))

cv.rectangle(img3, pt1=(354,0), pt2=(40,170), color=(0,255,0), thickness=10)
cv.rectangle(img3, pt1=(200,200), pt2=(510,150), color=(0,255,0), thickness=10)
cv.rectangle(img3, pt1=(250,300), pt2=(310,150), color=(0,255,0), thickness=10)
cv.rectangle(img3, pt1=(400,700), pt2=(510,150), color=(0,255,0), thickness=10)
cv.rectangle(img3, pt1=(500,700), pt2=(600,600), color=(0,255,0), thickness=10)

plt.imshow(img3)
```

```
# 5. Tampilkan image dengan posisi terbalik!
img5 = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB), 0)
plt.imshow(img5)
```

```
# 6. Buat rectangle dan circle pada bagian wajah dari image kitten!

# imgt = img
imgt = cv.resize(cv.cvtColor(img,cv.COLOR_BGR2RGB),(850,850))
cv.rectangle(imgt,pt1=(250,100),pt2=(500,300),color=(0,0,255), thickness=10)
cv.circle(imgt,center=(400,200),radius=100,color=(255,0,0), thickness=8)
plt.imshow(imgt)
```
